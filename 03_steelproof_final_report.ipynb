# Relatório da Solução

## 1. Descrição executiva

A Steelproof busca reduzir custos de produção por meio de um controle mais eficiente do consumo de energia durante o processamento do aço. Nesse contexto, o problema de negócio deste projeto foi prever a temperatura final do metal por lote, de forma a apoiar decisões operacionais mais precisas sobre aquecimento, tempo de processo e uso de insumos.

A relevância dessa análise está no fato de que a temperatura final influencia diretamente a eficiência energética, a estabilidade do processo e a qualidade do produto final. Um modelo preditivo confiável poderia ajudar a reduzir desperdícios, evitar aquecimento excessivo ou insuficiente e apoiar a tomada de decisão na planta industrial.

## 2. Descrição dos dados utilizados

Foram utilizados sete conjuntos de dados provenientes de diferentes etapas do processo siderúrgico:

- `data_arc_en.csv` — dados dos eletrodos e aquecimento;
- `data_bulk_en.csv` — volumes de materiais a granel;
- `data_bulk_time_en.csv` — tempos de adição de materiais a granel;
- `data_gas_en.csv` — dados do processo de purga de gás;
- `data_temp_en.csv` — medições de temperatura;
- `data_wire_en.csv` — volumes de materiais em arame;
- `data_wire_time_en.csv` — tempos de adição de materiais em arame.

A qualidade geral dos dados foi suficiente para viabilizar a modelagem, mas com limitações importantes. Entre os principais pontos observados estavam:
- colunas temporais inicialmente armazenadas como texto;
- alta esparsidade em várias colunas de materiais;
- necessidade de consolidar múltiplas fontes em uma única base por lote;
- presença de valores ausentes que, em muitos casos, representavam ausência de adição de material e não erro de coleta.

Essas limitações exigiram tratamento cuidadoso e influenciaram diretamente a estratégia de preparação dos dados e a interpretação dos resultados.

## 3. Descrição técnica da solução desenvolvida

A variável-alvo do projeto foi definida como a **temperatura final do metal por lote**, obtida a partir da última medição registrada para cada `key` em `data_temp`.

A solução foi construída em etapas:
- conversão de variáveis temporais para `datetime`;
- ordenação das medições de temperatura por lote;
- criação da temperatura inicial e da temperatura final;
- agregação dos dados de aquecimento elétrico por lote;
- integração dos dados de gás, materiais a granel e materiais em arame;
- criação de atributos adicionais relacionados a duração, potência, contagem de adições e volume total de materiais;
- tratamento contextual de valores ausentes;
- treinamento e comparação de modelos de regressão.

As principais métricas utilizadas foram:
- **MAE (Mean Absolute Error)** — métrica principal, por ser mais interpretável em unidades de temperatura;
- **RMSE (Root Mean Squared Error)** — para penalizar mais fortemente erros grandes;
- **R²** — para avaliar o poder explicativo global do modelo.

Os modelos testados foram:
- `DummyRegressor` como baseline;
- `LinearRegression`;
- `RandomForestRegressor`;
- `GradientBoostingRegressor`.

Também foi realizado ajuste de hiperparâmetros com `GridSearchCV` para os modelos baseados em ensemble.

## 4. Quais passos do plano foram executados e quais foram ajustados?

Os principais passos definidos no plano de trabalho foram executados:
- limpeza e consolidação dos dados por lote;
- criação da variável-alvo;
- construção de atributos preditores;
- treinamento de um modelo base;
- comparação entre múltiplos algoritmos;
- ajuste de hiperparâmetros;
- análise de importância das variáveis;
- interpretação dos resultados no contexto do negócio.

Nenhuma etapa central foi ignorada, mas alguns pontos foram ajustados ao longo do processo. Em especial, foi necessário incluir uma rodada adicional de engenharia de atributos e um experimento diagnóstico comparando split temporal e split aleatório. Essa adaptação não estava prevista de forma tão detalhada no plano inicial, mas se tornou necessária porque os primeiros resultados indicaram diferença significativa de desempenho entre validação e teste final.

## 5. Quais dificuldades foram encontradas e como foram resolvidas?

A principal dificuldade do projeto foi transformar múltiplas fontes operacionais em uma base analítica única por lote. Como os arquivos representavam diferentes etapas do processo e tinham granularidades distintas, foi necessário definir cuidadosamente como agregar cada informação.

Outra dificuldade importante foi a alta esparsidade em diversas colunas de materiais. Esse problema foi tratado com base no contexto do caso: em vez de assumir erro de coleta, os valores ausentes foram interpretados, quando apropriado, como ausência de adição de determinado material.

Também houve dificuldade para alcançar os limiares de desempenho planejados na avaliação temporal. Para investigar esse comportamento, foi realizada uma rodada adicional de engenharia de atributos e, posteriormente, um experimento diagnóstico com split aleatório. Essa comparação mostrou que os modelos aprendem sinal preditivo relevante, mas sofrem perda de desempenho ao generalizar para períodos futuros, o que sugere drift temporal no processo.

## 6. Quais foram as etapas principais de resolução da tarefa?

As etapas principais foram:

1. **Preparação e limpeza dos dados**  
   Conversão de datas, análise de valores ausentes, tratamento de colunas muito esparsas e consolidação das fontes por lote.

2. **Construção da base analítica**  
   Definição da temperatura final como target e criação de atributos relacionados a temperatura inicial, aquecimento elétrico, gás, materiais e tempo de processo.

3. **Treinamento e comparação de modelos**  
   Desenvolvimento de um baseline e de modelos de regressão mais robustos, seguido por ajuste de hiperparâmetros.

4. **Refinamento e diagnóstico**  
   Criação de atributos adicionais e comparação entre avaliação temporal e aleatória para entender melhor a capacidade de generalização dos modelos.

5. **Interpretação dos resultados**  
   Análise do melhor modelo, da importância das variáveis e das implicações dos resultados para o contexto operacional da Steelproof.

## 7. Qual é o modelo final e qual pontuação de qualidade ele tem?

Na avaliação principal do projeto, baseada em **split temporal**, o melhor modelo foi o **Random Forest ajustado com engenharia adicional de atributos**.

Seu desempenho foi:
- **MAE = 15.44**
- **RMSE = 19.60**
- **R² = 0.21**

Esse modelo superou o baseline e também apresentou o melhor equilíbrio entre erro absoluto médio e estabilidade entre os modelos testados no cenário temporal.

Além disso, no experimento diagnóstico com **split aleatório**, os resultados melhoraram substancialmente:
- **Random Forest ajustado:** MAE = 7.57, RMSE = 9.89, R² = 0.60
- **Gradient Boosting ajustado:** MAE = 7.47, RMSE = 9.74, R² = 0.62

Esses números mostram que o problema não está na ausência de sinal preditivo nos dados, mas sim na dificuldade de generalizar para períodos futuros, indicando mudança temporal no processo.

## 8. Estimativa de impacto no negócio

Do ponto de vista estratégico, o projeto mostrou que existe potencial real para apoiar a operação com modelos preditivos de temperatura, mas que a aplicação em produção exigiria atenção especial à estabilidade temporal do processo.

Mesmo sem estimar financeiramente um valor exato, é possível afirmar qualitativamente que um modelo com melhor generalização poderia gerar impacto positivo ao:
- reduzir consumo excessivo de energia em aquecimentos desnecessários;
- diminuir retrabalho operacional causado por temperaturas fora da faixa desejada;
- aumentar previsibilidade do processo;
- apoiar decisões mais consistentes sobre aquecimento e adição de insumos.

Ao mesmo tempo, os resultados sugerem que, para gerar valor confiável em ambiente real, o modelo precisaria ser monitorado e atualizado periodicamente, especialmente se houver mudanças operacionais ao longo do tempo.

## 9. Principais conclusões e recomendações

A principal conclusão do projeto é que os dados disponíveis permitem construir modelos que aprendem relações relevantes entre variáveis de processo e temperatura final do metal. No entanto, a forte diferença entre o desempenho em split temporal e em split aleatório indica que a generalização do modelo é afetada por **drift temporal**, ou seja, por mudanças de padrão ao longo do tempo.

As variáveis mais relevantes para o melhor modelo incluíram:
- temperatura inicial;
- contagem de adições em arame;
- soma de potência ativa;
- duração total de aquecimento;
- algumas variáveis específicas de materiais.

Com base nisso, as principais recomendações são:
1. monitorar mudanças de regime operacional ao longo do tempo;
2. considerar re-treinamento periódico do modelo com dados mais recentes;
3. investigar variáveis adicionais que possam explicar melhor o balanço térmico do processo;
4. usar a solução atual como base analítica para evolução, e não como sistema final definitivo sem manutenção.

Em resumo, o projeto gerou uma solução tecnicamente consistente e revelou um achado importante para o negócio: a precisão do modelo depende não apenas dos atributos disponíveis, mas também da estabilidade temporal do processo produtivo.
